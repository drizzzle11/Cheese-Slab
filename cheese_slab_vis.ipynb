#### Step 1: Read the Parquet file and convert to DataFrame

```python
# Import relevant libraries
import pandas as pd
import os
import matplotlib.pyplot as plt

# Parquet file 
all_elements = 'C:\\Users\\rimjhims\\Walter P. Moore and Associates\\S17-24001-00 DB Texas Instruments LFAB2 - CheeseSlab\\Cheese Slab Export\\all_elements.parquet'
```

```python
# Covert to DataFrame
all_elements_df = pd.read_parquet(all_elements)
```

#### Step 2: Plot the points using matplotlib

```python
all_elements_df.columns
```

```python
# Extracting values for plotting
x = all_elements_df['CentroidX']
y = all_elements_df['CentroidY']
colors = all_elements_df['min_P']

# Create scatter plot with 'magma' colormap
plt.figure(figsize=(10, 6))
scatter = plt.scatter(x, y, c=colors, cmap='magma', s=5)

# Add colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('min_P Value')

# Add labels and title
plt.xlabel('CentroidX')
plt.ylabel('CentroidY')
plt.title('Scatter Plot of CentroidX vs CentroidY with min_P as Color')

# Add grid for better readability
plt.grid(True)

# Show the plot
plt.show()
```
